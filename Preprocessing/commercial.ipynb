```python
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
```

```python
commercial = pd.read_csv('/content/drive/MyDrive/Dataset/Origin/commercial_dataset.csv', encoding='cp949')
print(commercial.columns)
```

```python
#2025년 4분기 데이터 추출
commercial = commercial[
    commercial['기준_년분기_코드'] == 20254
]
```

```python
#medical
commercial['medical'] = commercial[
    ['종합병원_수','일반_병원_수','약국_수']
].sum(axis=1)

#education
commercial['education'] = commercial[
    ['유치원_수','초등학교_수','중학교_수',
     '고등학교_수','대학교_수']
].sum(axis=1)

#shopping
commercial['shopping'] = commercial[
    ['백화점_수','슈퍼마켓_수']
].sum(axis=1)

#bank
commercial['bank'] = commercial['은행_수']

#gov
commercial['gov'] = commercial['관공서_수']

#culture
commercial['culture'] = commercial[
    ['극장_수','숙박_시설_수']
].sum(axis=1)

#intercity
commercial['intercity'] = commercial[
    ['공항_수','철도_역_수']
].sum(axis=1)

#intracity
commercial['intracity'] = commercial[
    ['버스_터미널_수',
     '지하철_역_수',
     '버스_정거장_수']
].sum(axis=1)
```

```python
commercial = commercial[
    [
        '행정동_코드',
        '행정동_코드_명',
        'medical',
        'education',
        'shopping',
        'bank',
        'gov',
        'culture',
        'intercity',
        'intracity'
    ]
]

commercial = commercial.fillna(0)
commercial.head()
```

```python
commercial.to_csv('/content/drive/MyDrive/Dataset/Preprocessed/commercial_group.csv', index=False)
```
